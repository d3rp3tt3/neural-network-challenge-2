## Part 1: Preprocessing

```python
# Import our dependencies
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
import pandas as pd
import numpy as np
from tensorflow.keras.models import Model
from tensorflow.keras import layers

#  Import and read the attrition data
attrition_df = pd.read_csv('https://static.bc-edx.com/ai/ail-v-1-0/m19/lms/datasets/attrition.csv')
attrition_df.head()
```

```python
# Determine the number of unique values in each column.
attrition_df.nunique()
```

```python
# Create y_df with the Attrition and Department columns
y_df = attrition_df[["Attrition", "Department"]]
y_df.head()
```

```python
# Create a list of at least 10 column names to use as X data

# Create X_df using your selected columns
X_df = attrition_df[['Age', 'Education', 'DistanceFromHome', 'JobSatisfaction', 'OverTime', 'StockOptionLevel', 'WorkLifeBalance', 'YearsAtCompany', 'YearsSinceLastPromotion', 'NumCompaniesWorked']]

# Show the data types for X_df
X_df.dtypes
```

```python
# Split the data into training and testing sets
from sklearn.model_selection import train_test_split
X_df_train, X_df_test, y_df_train, y_df_test = train_test_split(X_df, y_df, random_state=42)
```

```python
# Get value counts for original encoded OverTime column
X_df_train['OverTime'].value_counts()
```

```python
# Convert your X data to numeric data types however you see fit
# Add new code cells as necessary
def convert_to_numeric(df, column_name):
    df[column_name] = df[column_name].replace({'Yes': 1, 'No': 0})
    return df
```

```python
# Apply the function to the OverTime column for the training data
X_df_train = convert_to_numeric(X_df_train, 'OverTime')
X_df_train['OverTime'].value_counts()
```

```python
# Verify encoding changes
X_df_train.head()
```

```python
# Get value counts for original encoded OverTime column
X_df_test['OverTime'].value_counts()
```

```python
# Apply the function to the OverTime column for the testing data
X_df_test = convert_to_numeric(X_df_test, 'OverTime')
X_df_test['OverTime'].value_counts()
```

```python
# Verify encoding changes
X_df_train.head()
```

```python
# Verify data types changes for training and testing data
print(X_df_train.dtypes)
print(X_df_test.dtypes)
```

```python
# Create a StandardScaler and fit it to the training data
scaler = StandardScaler().fit(X_df_train)

# Scale the training and testing data
X_train_scaled = scaler.transform(X_df_train)
X_test_scaled = scaler.transform(X_df_test)
```

```python
# Create a OneHotEncoder for the Department column
from sklearn.preprocessing import OneHotEncoder
dept_encoder = OneHotEncoder(handle_unknown='infrequent_if_exist', sparse_output=False)

# Fit the encoder to the training data
dept_encoded = dept_encoder.fit(y_df_train[['Department']])

# Create two new variables by applying the encoder to the training and testing data
y_train_encoded_dept = dept_encoded.transform(y_df_train[['Department']])
y_test_encoded_dept = dept_encoded.transform(y_df_test[['Department']])
y_train_encoded_dept
```

```python
# Create a OneHotEncoder for the Attrition column
attr_encoder = OneHotEncoder(handle_unknown='infrequent_if_exist', sparse_output=False)

# Fit the encoder to the training data
attr_encoded = attr_encoder.fit(y_df_train[['Attrition']])

# Create two new variables by applying the encoder to the training and testing data
y_train_encoded_attr = attr_encoded.transform(y_df_train[['Attrition']])
y_test_encoded_attr = attr_encoded.transform(y_df_test[['Attrition']])
y_train_encoded_attr
```

## Create, Compile, and Train the Model

```python
# Find the number of columns in the X training data
print(X_train_scaled.shape)

# Create the input layer
input = layers.Input(shape=(10,), name='input_layer')

# Create at least two shared layers
shared1 = layers.Dense(64, activation='relu', name='shared1')(input)
shared2 = layers.Dense(128, activation='relu', name='shared2')(shared1)
```

```python
# Create a branch for Department with a hidden layer and an output layer

# Create the hidden layer
hidden_layer_dept = layers.Dense(units=32, name='department_hidden', activation='relu')(shared2)

# Create the output layer
dept_output_layer = layers.Dense(y_train_encoded_dept.shape[1],
                             activation='sigmoid',
                             name='department_output')(hidden_layer_dept)
```

```python
# Create a branch for Attrition with a hidden layer and an output layer

# Create the hidden layer
hidden_layer_att = layers.Dense(units=32, name='attrition_hidden', activation='relu')(shared2)

# Create the output layer
att_output_layer = layers.Dense(y_train_encoded_attr.shape[1],
                             activation='sigmoid',
                             name='attrition_output')(hidden_layer_att)
```

```python
# Create the model
from tensorflow.keras.metrics import BinaryAccuracy, CategoricalAccuracy, Precision, Recall, AUC
model = Model(inputs=input, outputs=[dept_output_layer, att_output_layer], name='attrition_model')

# Compile the model
model.compile(optimizer='adam',
              loss={
                  'department_output': 'categorical_crossentropy',
                  'attrition_output': 'binary_crossentropy'
              },
              metrics={
                  'department_output': [
                      'accuracy',
                      Precision(name='precision_dept'),
                      Recall(name='recall_dept'),
                      AUC(name='auc_dept')
                  ],
                  'attrition_output': [
                      'accuracy',
                      Precision(name='precision_attr'),
                      Recall(name='recall_attr'),
                      AUC(name='auc_attr')
                  ]
              })

# Summarize the model
model.summary()
```

```python
# Train the model
model.fit(X_train_scaled, {'department_output': y_train_encoded_dept, 'attrition_output': y_train_encoded_attr},
          epochs=100, shuffle=True, verbose=2)
```

```python
from tensorflow.keras import backend as K
from sklearn.metrics import roc_auc_score

# Assuming X_test_scaled, y_test_encoded_dept, and y_test_encoded_attr are already prepared

# Function to calculate AUC for binary classification
def binary_auc(y_true, y_pred):
    return roc_auc_score(y_true, y_pred)

# Ensure y_test_encoded_dept and y_test_encoded_attr are numpy arrays
y_test_encoded_dept = np.array(y_test_encoded_dept)
y_test_encoded_attr = np.array(y_test_encoded_attr)

# Evaluate the model with the testing data
results = model.evaluate(
    np.array(X_test_scaled),
    {
        'department_output': y_test_encoded_dept,
        'attrition_output': y_test_encoded_attr
    },
    verbose=2
)

print("Model Performance:")
for i, metric in enumerate(model.metrics_names):
    print(f"{metric}: {results[i]}")

# Calculate additional metrics
y_pred_dept = model.predict(np.array(X_test_scaled))['department_output']
y_pred_attr = model.predict(np.array(X_test_scaled))['attrition_output']

print("\nAdditional Metrics:")
print(f"Attrition Output AUC: {binary_auc(y_test_encoded_attr, y_pred_attr)}")
print(f"Department Output AUC: {binary_auc(y_test_encoded_dept, y_pred_dept)}")
print(f"Attrition Output Precision: {precision(y_test_encoded_attr, y_pred_attr)}")
print(f"Attrition Output Recall: {recall(y_test_encoded_attr, y_pred_attr)}")
print(f"Department Output Precision: {precision(y_test_encoded_dept, y_pred_dept)}")
print(f"Department Output Recall: {recall(y_test_encoded_dept, y_pred_dept)}")
```

```python
# Print the accuracy for both department and attrition
print(f"Department Accuracy %: {results[1]*100}")
print(f"Attrition Accuracy %: {results[2]*100}")
```
